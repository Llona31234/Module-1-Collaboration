```python
import multiprocessing
import random
import time
from datetime import datetime

def worker(process_id):
    delay = random.random()
    time.sleep(delay)

    current_time = datetime.now().strftime("%H:%M:%S.%f")

    print(f"Process {process_id} finished at {current_time} after {delay:.3f} seconds")


if __name__ == "__main__":
    processes = []

    for i in range(3):
        p = multiprocessing.Process(target=worker, args=(i,))
        processes.append(p)
        p.start()

    for p in processes:
        p.join()

    print("All processes completed.")
```
