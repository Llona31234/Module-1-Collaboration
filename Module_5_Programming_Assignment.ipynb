## 1. Testing Your Code

Testing is used to check if a program works correctly. There are two main types of testing: manual testing and automated testing.

Manual testing is when a person runs the program and checks by trying different inputs. This is also called exploratory testing. However, it can be repetitive.

Automated testing uses scripts to run tests automatically. This is faster and more reliable, especially for large programs.

There are also two important testing types:
Unit testing: tests small parts of a program
Integration testing: tests multiple parts working together

Unit testing is useful because it helps find exactly where an error occurs.

## 2. Writing Your First Test

To write a test, you first create a function called sum() inside a module. This function takes a list of numbers and returns their total. Then I created a separate test file to test the function.

There are three main steps that the test follows. The first one is creating input data. The second one is running the function. The third one is comparing the result with the expected output. I used assertEqual() to check if the result matches the expected value. I also learned that tests should be repeatable, including cases like negative numbers or invalid inputs.

## 3. Executing Your First Test

After writing the test, I ran it using the unittest test runner in VS Code. The test runner runs all test cases and checks if they pass or fail. A dot (.) means the test passed, while an F means the test failed.

I also used certain commands such as python test.py and python -m unittest test. Verbose mode shows more details, including test names and results.

When a test fails, the output shows the test name, the line where the error occurred, and, lastly, the expected and actual results. This makes it easier to find and fix problems in the code.

## 4. Test Results

The test results of my code are working correctly. If a test passes, it means the function returned the expected result. If a test fails, it means there is a difference between the expected and actual output. The results also provide useful debugging information, such as the name of the failed test, the line of code where the error occurred, and the difference between the expected and actual values.
